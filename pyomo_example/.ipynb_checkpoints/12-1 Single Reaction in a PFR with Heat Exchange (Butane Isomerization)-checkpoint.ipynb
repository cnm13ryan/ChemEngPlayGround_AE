```python
from pyomo.environ import *
from pyomo.dae import *

# Create a model
m = ConcreteModel()

# Define the parameters
m.k0 = Param(initialize=31.1)  # Pre-exponential factor in h⁻¹
m.Ea = Param(initialize=65.7e3)  # Activation energy in J/mol
m.R = Param(initialize=8.314)  # Universal gas constant in J/mol·K
m.Cp_A = Param(initialize=141e3)  # Heat capacity of A in J/mol·K
m.Cp_0 = Param(initialize=159e3)  # Total heat capacity in J/mol·K
m.deltaH = Param(initialize=-34500e3)  # Heat of reaction in J/mol
m.U = Param(initialize=5000e3/3600)  # Heat transfer coefficient in J/m³ s K
m.C_A0 = Param(initialize=1.86)  # Initial concentration of A in mol/L
m.T0 = Param(initialize=305)  # Initial temperature in K
m.Tc_in = Param(initialize=315)  # Coolant inlet temperature in K
m.CpC = Param(initialize=28e3)  # Heat capacity of coolant in J/kg·K
m.F0 = Param(initialize=14.7)  # Molar flow rate in kmol/h

# Define the differential variables
m.V = ContinuousSet(bounds=(0, 5))  # Reactor volume from 0 to 5 m³
m.X = Var(m.V, within=NonNegativeReals, bounds=(0, 1))  # Conversion
m.T = Var(m.V, within=NonNegativeReals)  # Temperature
m.Tc = Var(m.V, within=NonNegativeReals)  # Coolant temperature

# Define the derivatives
m.dX_dV = DerivativeVar(m.X, wrt=m.V)
m.dT_dV = DerivativeVar(m.T, wrt=m.V)
m.dTc_dV = DerivativeVar(m.Tc, wrt=m.V)

# Define the equations

# Mole balance for A
def mole_balance_rule(model, V):
    return m.F0 * m.dX_dV[V] == -m.k0 * exp(-m.Ea / (m.R * m.T[V])) * m.C_A0 * (1 - m.X[V])
m.mole_balance = Constraint(m.V, rule=mole_balance_rule)

# Energy balance for the reactor
def energy_balance_rule(model, V):
    return m.F0 * m.Cp_0 * m.dT_dV[V] == m.k0 * exp(-m.Ea / (m.R * m.T[V])) * m.C_A0 * (1 - m.X[V]) * m.deltaH + m.U * (m.Tc[V] - m.T[V])
m.energy_balance = Constraint(m.V, rule=energy_balance_rule)

# Energy balance for the coolant
def coolant_energy_balance_rule(model, V):
    return m.dTc_dV[V] == -m.U * (m.Tc[V] - m.T[V]) / m.CpC
m.coolant_energy_balance = Constraint(m.V, rule=coolant_energy_balance_rule)

# Initial conditions
m.X[0].fix(0)  # Conversion is 0 at V=0
m.T[0].fix(m.T0)  # Temperature is T0 at V=0
m.Tc[0].fix(m.Tc_in)  # Coolant temperature is Tc_in at V=0

# Solve the model
solver = SolverFactory('ipopt')
solver.solve(m, tee=True)

# Post-processing and plotting can be done using matplotlib or any other preferred library

```

```python
# Post-process the results and plot

V_values = [v for v in m.V]
X_values = [m.X[v].value for v in m.V]
T_values = [m.T[v].value for v in m.V]
T_coolant_values = [m.T_coolant[v].value for v in m.V]

# Plot Conversion (X) vs. Reactor Volume
plt.figure(figsize=(10, 6))
plt.plot(V_values, X_values, label='Conversion (X)', color='blue')
plt.xlabel('Reactor Volume (m^3)')
plt.ylabel('Conversion (X)')
plt.title('Conversion Profile along the Reactor Length')
plt.grid(True)
plt.legend()
plt.show()

# Plot Temperature (T) vs. Reactor Volume
plt.figure(figsize=(10, 6))
plt.plot(V_values, T_values, label='Temperature (T)', color='red')
plt.xlabel('Reactor Volume (m^3)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Profile along the Reactor Length')
plt.grid(True)
plt.legend()
plt.show()

# Plot Coolant Temperature (T_coolant) vs. Reactor Volume
plt.figure(figsize=(10, 6))
plt.plot(V_values, T_coolant_values, label='Coolant Temperature (T_coolant)', color='green')
plt.xlabel('Reactor Volume (m^3)')
plt.ylabel('Coolant Temperature (K)')
plt.title('Coolant Temperature Profile along the Reactor Length')
plt.grid(True)
plt.legend()
plt.show()
```
